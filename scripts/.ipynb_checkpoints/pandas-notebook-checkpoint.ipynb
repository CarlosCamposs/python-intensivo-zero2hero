## Filtrado por condiciones
Se busca filtrar filas cuyas algunas de sus columnas se cumpla alguna condición lógica

```python
df[df["favorites"]>400]
```

```python
#####
# Para usar mas de dos condicionales, se usa el símbolo & y se deben encerrar entre paréntesis

df[ (df["favorites"]>400) & (df["mentions"]>20)]

# El símbolo para el "or" es |
```

```python
##### Para buscar alguna string

df[df["full_text"].str.contains("Programming")]
        # Busca todos aquellos registros cuya columna full_text contenga la palabra programming
```

## Transformación de datos
A partir de datos ya existentes, se busca crear nuevos. Primero se debe definir la función que va a transformar los datos

```python
import random

def calcularGanancias(x):
    ganancia=x*random.randint(3,5)
    return ganancia

# Para crear la nueva columna del data frame

df["ganancias"]=df["retweets"].apply(calcularGanancias)
            # Lo que dice la sentencia se puede leer como:
            # "Por cada valor de la columna retweets, aplícale la funcion calcularGanancias"
        
df.head()
```

Lo siguiente que veremos es como calcular una nueva columna a partir de los valores de **varias** columnas del df

```python
def popularidad(x):
    resultado=x["followees"]/x["followers"]
    return resultado

df["popularidad"]=df.apply(popularidad,axis=1)
    # axis=1 especifica que la función debe ser aplicada a cada fila del df
    # (anteriormente se aplicaba a solo una columna)
    
df.head()
```
