# Manipulación de data frames

```python
df=pd.DataFrame({
    'key':['A','B','C','A','B','C'],
    'value1':range(6),
    'value2':np.random.randint(0,100,6) # Queremos "6" numeros aleatorios enteros entre 0 y 1000
})
```

```python
df
```

## Aggregate

```python
df.groupby('key').aggregate(['min','max',np.median, np.mean, max])
```

```python
df.value1.aggregate([min,max,np.mean]) # Esta calculando TRES estadisticos de una misma columna
```

```python
df.groupby('key').aggregate({
    'value1':min,
    'value2':max
})
```

## Filter

```python
df.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std()>17 # Esto es un valor booleano 
```

```python
df.groupby('key').filter(filter_func)
    # Como la categoria B no cumple la condición que std>17, entonces en el resultado de esta linea
    # no se inlcuyen los registros de la key con valor "B", pero muestra los datos originales, es decir, no los agrupados
```

```python
# -----------------------------------------------------------------------------------------------------------------------------
# Otro ejemplo, mejor ejemplo apra entender filter
df.groupby('key').aggregate([np.mean,np.std])
```

```python
def comparar_mean(x):
    return x['value1'].mean()>2.5 and x['value2'].std()<58
```

```python
df.groupby('key').filter(comparar_mean)
    # Solo la categoria "C" cumple ambos criterios
# -----------------------------------------------------------------------------------------------------------------------------
```

## Transform

```python
df.groupby('key').transform(lambda x: x-x.mean())
    # Calcula la media de la clase A, B y C y a cada valor de la tabal original le resta la media de la clase a la que pertence
    # -1.5 = 0 - 1.5
    # ie, x.mean() se refiere a la media de cada clase
```

```python
# -----------------------------------------------------------------------------------------------------------------------------
# Mejor ejemplo para explicar el trasnform
df.value1 # Mostramos los datos
```

```python
df.value1.transform(lambda x: x-5) # Aplicamos un transform
# ---------------------------------------------------------------------------------------------------------------------------
```

##  Apply

```python
def norm(x):
    x['value1']/=x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm)
```
