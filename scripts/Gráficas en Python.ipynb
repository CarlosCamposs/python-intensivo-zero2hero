# Gr치ficas en Python

```python
 import matplotlib.pyplot as plt
    
# Para instalar una biblioteca: pip install matplotlib.pyplot
```

#### Variables dependientes e independientes

```python
dominio=range(100)
```

#### Relaci칩n

```python
def rel1(x):
    return x**3-(3*x**2)

def rel2(x):
    return x**3
```

```python
# Generamos los valores del rango

rango1=[]
rango2=[]

for x in dominio:
    rango1.append(rel1(x))
    rango2.append(rel2(x))
```

#### La gr치fica m치s sencilla del mundo

```python
plt.plot(dominio, rango1)
```

#### Figure

```python
# Para ponerle mas detalle, tenemos que crear una "figure"
plt.figure(figsize=(12,8))

plt.plot(dominio,rango1,'g',label='rel1') # "g" es de green
plt.plot(dominio,rango2,'r--',label='rel2') # "r" es de red; "--" es para puntear

plt.title('Gr치fica 1: Relaci칩n')

plt.xlabel('Dominio')
plt.ylabel('Rango')

plt.legend() # Con esto permite poner las label "rel1" y "rel2"
plt.show() # Esto har치 que se muestre autom치ticamente
```

#### Numpy

```python
import numpy as np
# Instalar biblioteca numpy: pip install numpy
```

```python
dominio=np.linspace(0,2*np.pi,100)

# linspace sirve para crear una secuencia de valores, entre 0 y 2*np.pi, y es una secuencia que tiene 100 valores
```

```python
def seno(x):
    return np.sin(x)

def coseno(x):
    return np.cos(x)
```

```python
rango1=[]
rango2=[]

# Creamos los valores del rango
for i in dominio:
    rango1.append(seno(i))
    rango2.append(coseno(i))
```

```python
plt.figure(figsize=(8,8))

plt.title('SENO VS COSENO')
plt.plot(dominio, rango1, '--', label='seno')
plt.plot(dominio, rango2,  label='coseno')
plt.legend()
plt.show()
```
